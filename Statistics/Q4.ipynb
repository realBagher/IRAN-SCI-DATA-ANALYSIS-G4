```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Rural
R98 = pd.read_excel('R98.xlsx', sheet_name='R98P3S11')
R99 = pd.read_excel('R99.xlsx', sheet_name='R99P3S11')
R1400 = pd.read_excel('R1400.xlsx', sheet_name='R1400P3S11')
R1401 = pd.read_excel('R1401.xlsx', sheet_name='R1401P3S11')
#Urban
U98 = pd.read_excel('U98.xlsx', sheet_name='U98P3S11')
U99 = pd.read_excel('U99.xlsx', sheet_name='U99P3S11')
U1400 = pd.read_excel('U1400.xlsx', sheet_name='U1400P3S11')
U1401 = pd.read_excel('U1401.xlsx', sheet_name='U1401P3S11')
```

```python
dfs_R = {'R98': R98, 'R99': R99, 'R1400': R1400, 'R1401': R1401}
dfs_U = {'U98': U98, 'U99': U99, 'U1400': U1400, 'U1401': U1401}
```

```python
for df in dfs_R.values():
    df['main_code'] = df['code'].astype('str').str[:4].astype('int')
    
for df in dfs_U.values():
    df['main_code'] = df['code'].astype('str').str[:4].astype('int')
```

```python
GroupData_1111_R = {}
GroupData_1112_R = {}
GroupData_1120_R = {}
R_ovrall = {}

for name, df in dfs_R.items():
    GroupData_1111_R[name] = df[df['main_code'] == 1111]['value'].sum()
    GroupData_1112_R[name] = df[df['main_code'] == 1112]['value'].sum()
    GroupData_1120_R[name] = df[df['main_code'] == 1120]['value'].sum()
    R_ovrall[name] = df['value'].sum()
```

```python
GroupData_1111_U = {}
GroupData_1112_U = {}
GroupData_1120_U = {}
U_ovrall = {}

for name, df in dfs_U.items():
    GroupData_1111_U[name] = df[df['main_code'] == 1111]['value'].sum()
    GroupData_1112_U[name] = df[df['main_code'] == 1112]['value'].sum()
    GroupData_1120_U[name] = df[df['main_code'] == 1120]['value'].sum()
    U_ovrall[name] = df['value'].sum()
```

```python
lens_1111 = {
    'U98' : U98[U98['main_code'] == 1111].nunique()['Address'],
    'U99' : U99[U99['main_code'] == 1111].nunique()['Address'],
    'U1400' : U1400[U1400['main_code'] == 1111].nunique()['Address'],
    'U1401' : U1401[U1401['main_code'] == 1111].nunique()['Address'],
    'R98' : R98[R98['main_code'] == 1111].nunique()['Address'],
    'R99' : R99[R99['main_code'] == 1111].nunique()['Address'],
    'R1400' : R1400[R1400['main_code'] == 1111].nunique()['Address'],
    'R1401' : R1401[R1401['main_code'] == 1111].nunique()['Address']
}
```

```python
years = ['98', '99', '1400', '1401']
```

```python
plot_Datas_1111 = {}

for year in years :
    plot_Datas_1111[year] = ((GroupData_1111_R['R' + year] * lens_1111['R' + year]) + (GroupData_1111_U['U' + year] * lens_1111['U' + year])) / (lens_1111['R' + year] + lens_1111['U' + year])
```

```python
lens_1112 = {
    'U98' : U98[U98['main_code'] == 1112].nunique()['Address'],
    'U99' : U99[U99['main_code'] == 1112].nunique()['Address'],
    'U1400' : U1400[U1400['main_code'] == 1112].nunique()['Address'],
    'U1401' : U1401[U1401['main_code'] == 1112].nunique()['Address'],
    'R98' : R98[R98['main_code'] == 1112].nunique()['Address'],
    'R99' : R99[R99['main_code'] == 1112].nunique()['Address'],
    'R1400' : R1400[R1400['main_code'] == 1112].nunique()['Address'],
    'R1401' : R1401[R1401['main_code'] == 1112].nunique()['Address']
}
```

```python
plot_Datas_1112 = {}

for year in years :
    plot_Datas_1112[year] = ((GroupData_1112_R['R' + year] * lens_1112['R' + year]) + (GroupData_1112_U['U' + year] * lens_1112['U' + year])) / (lens_1112['R' + year] + lens_1112['U' + year])
```

```python
lens_1120 = {
    'U98' : U98[U98['main_code'] == 1120].nunique()['Address'],
    'U99' : U99[U99['main_code'] == 1120].nunique()['Address'],
    'U1400' : U1400[U1400['main_code'] == 1120].nunique()['Address'],
    'U1401' : U1401[U1401['main_code'] == 1120].nunique()['Address'],
    'R98' : R98[R98['main_code'] == 1120].nunique()['Address'],
    'R99' : R99[R99['main_code'] == 1120].nunique()['Address'],
    'R1400' : R1400[R1400['main_code'] == 1120].nunique()['Address'],
    'R1401' : R1401[R1401['main_code'] == 1120].nunique()['Address']
}
```

```python
plot_Datas_1120 = {}

for year in years :
    plot_Datas_1120[year] = ((GroupData_1120_R['R' + year] * lens_1120['R' + year]) + (GroupData_1120_U['U' + year] * lens_1120['U' + year])) / (lens_1120['R' + year] + lens_1120['U' + year])
```

```python
lens = {
    'U98' : U98.nunique()['Address'],
    'U99' : U99.nunique()['Address'],
    'U1400' : U1400.nunique()['Address'],
    'U1401' : U1401.nunique()['Address'],
    'R98' : R98.nunique()['Address'],
    'R99' : R99.nunique()['Address'],
    'R1400' : R1400.nunique()['Address'],
    'R1401' : R1401.nunique()['Address']
}
```

```python
plot_Datas = {}
for year in years :
    plot_Datas[year] = (((GroupData_1111_R['R' + year] + GroupData_1112_R['R' + year] + GroupData_1120_R['R' + year]) * lens['R' + year]) + \
                        ((GroupData_1111_U['U' + year] + GroupData_1112_U['U' + year] + GroupData_1120_U['U' + year]) * lens['U' + year])) / (lens['R' + year] + lens['U' + year])
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(GroupData_1111_R.keys(), GroupData_1111_R.values(), 
        marker='o', label='Meal Cost Rural', linestyle='-', linewidth=3)
ax.plot(GroupData_1112_R.keys(), GroupData_1112_R.values(), 
        marker='^', label='Cafeterias Rural', linestyle='--', linewidth=3)
ax.plot(GroupData_1120_R.keys(), GroupData_1120_R.values(), 
        marker='x', label='Accommodation Services Rural', linestyle=':', linewidth=3)
ax.plot(R_ovrall.keys(), R_ovrall.values(), 
        marker='v', label='Whole Data Rural', linestyle='-.', linewidth=3)

ax.set_title('Cost Analysis Over the Years of Rural Base on 14829 Questionnaire', fontsize=18, fontweight='bold')
ax.set_xlabel('Years', fontsize=14)
ax.set_ylabel('Values', fontsize=14)

ax.grid(True)
ax.legend(fontsize=12)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(GroupData_1111_U.keys(), GroupData_1111_U.values(), 
        marker='o', label='Meal Cost Urban', linestyle='-', linewidth=3)
ax.plot(GroupData_1112_U.keys(), GroupData_1112_U.values(), 
        marker='^', label='Cafeterias Urban', linestyle='--', linewidth=3)
ax.plot(GroupData_1120_U.keys(), GroupData_1120_U.values(), 
        marker='x', label='Accommodation Services Urban', linestyle=':', linewidth=3)
ax.plot(U_ovrall.keys(), U_ovrall.values(), 
        marker='v', label='Whole Data Rural ', linestyle='-.', linewidth=3)

ax.set_title('Cost Analysis Over the Years of Urban Base on 22703 Questionnaire', fontsize=18, fontweight='bold')
ax.set_xlabel('Years', fontsize=14)
ax.set_ylabel('Values', fontsize=14)

ax.grid(True)
ax.legend(fontsize=12)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(plot_Datas_1111.keys(), plot_Datas_1111.values(), 
        marker='o', label='Meal Cost', linestyle='-', linewidth=3)
ax.plot(plot_Datas_1112.keys(), plot_Datas_1112.values(), 
        marker='^', label='Cafeterias', linestyle='--', linewidth=3)
ax.plot(plot_Datas_1120.keys(), plot_Datas_1120.values(), 
        marker='x', label='Accommodation Services', linestyle=':', linewidth=3)
ax.plot(plot_Datas.keys(), plot_Datas.values(), 
        marker='v', label='Whole Data', linestyle='-.', linewidth=3)

ax.set_title('Cost Analysis Over the Years Base on 37532 Questionnaire', fontsize=18, fontweight='bold')
ax.set_xlabel('Years', fontsize=14)
ax.set_ylabel('Values', fontsize=14)

ax.grid(True)
ax.legend(fontsize=12)

plt.show()
```
