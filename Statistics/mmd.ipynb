```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_r_1401 = pd.read_excel('R1401.xlsx', sheet_name='R1401P1')
data_u_1401 = pd.read_excel('U1401.xlsx', sheet_name='U1401P1')
```

```python
100 * (data_u_1401.isna().sum(axis=0) / data_u_1401.shape[0])
```

```python
degree_map = {'Elemantry': 0,
'Secondary': 1,
'HighSchool': 2,
'Diploma': 3,
'College': 4,
'Bachelor': 5,
'Master': 6,
'PhD': 7,
'Other': 8
}
degree = ['Elemantry',
'Secondary',
'HighSchool',
'Diploma',
'College',
'Bachelor',
'Master',
'PhD',
'Other']
relation_map = {
    'Head': 1,
    'Spouse': 2,
    'Child': 3,
    'SonDaughter_inLaw': 4,
    'GrandSonDaughter': 5,
    'Parent': 6,
    'Sibling': 7,
    'OtherRelative': 8,
    'NonRelative': 9
}
relation = ['Head',
'Spouse',
'Child',
'SonDaughter_inLaw',
'GrandSonDaughter',
'Parent',
'Sibling',
'OtherRelative',
'NonRelative']
```

```python
data_r_1401['degree_enc'] = data_r_1401['degree'].map(degree_map)
data_r_1401['relation_enc'] = data_r_1401['relation'].map(relation_map)
```

```python
data_u_1401['degree_enc'] = data_u_1401['degree'].map(degree_map)
data_u_1401['relation_enc'] = data_u_1401['relation'].map(relation_map)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(25, 8))
fig.suptitle('1401')
axes = axes.flatten()
data_r_1401['age'].hist(bins=20,ax=axes[0])
axes[0].set_title('Age')
data_r_1401['degree_enc'].hist(bins=9, ax=axes[1], rwidth=0.8)
axes[1].set_title('Degree')
axes[1].set_xticks(range(len(degree)))
axes[1].set_xticklabels(degree, rotation=30)

data_r_1401['relation_enc'].hist(bins=9, ax=axes[2], rwidth=0.9)
axes[2].set_title('Relation')
axes[2].set_xticks(range(len(relation)))
axes[2].set_xticklabels(relation, rotation=90)


data_u_1401['age'].hist(bins=20,ax=axes[3], alpha = 0.5)
data_u_1401['degree_enc'].hist(bins=9, ax=axes[4], rwidth=0.8, alpha = 0.5)
axes[4].set_xticks(range(len(degree)))
axes[4].set_xticklabels(degree, rotation=30)

data_u_1401['relation_enc'].hist(bins=9, ax=axes[5], rwidth=0.9, alpha = 0.5)
axes[5].set_xticks(range(len(relation)))
axes[5].set_xticklabels(relation, rotation=90)
plt.show()
```
